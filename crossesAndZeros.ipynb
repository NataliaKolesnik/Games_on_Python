```python
L0 = [" ", '1', '2', '3']
L1 = ['1', "-", "-", "-"]
L2 = ['2', "-", "-", "-"]
L3 = ['3', "-", "-", "-"]


def error_coordinates(coordinate):
    """Данная функция осуществляет проверку координат на корректность ввода
    в случае корректного вода возвращает координаты в форме списка, иначе False
    """
    
    #Проверка, что введены только числа
    if coordinate.replace(' ','').isdigit():
        coordinate = list(map(int, coordinate.split())) 
    else:
        print('Введите координаты числами! от 1 до 3')
        return False
    
    #Проверка на количество координат - их должно быть 2
    if len(coordinate) != 2:
        print('Вы ввели неверное колличество координат. Повторите ввод.')
        return False
    #Если количество координат верно, то проверка свободно ли место
    else: 
        if 0 < coordinate[0] < 4 and 0 < coordinate[1] < 4:
            if coordinate[0] == 1:
                if L1[coordinate[1]] != "-":
                    print(f"Место '{coordinate[0]} {coordinate[1]}' уже занято. Повторите ввод.")
                    return False
            elif coordinate[0] == 2:
                if L2[coordinate[1]] != "-":
                    print(f"Место '{coordinate[0]} {coordinate[1]}' уже занято. Повторите ввод.")
                    return False
            elif coordinate[0] == 3:
                if L3[coordinate[1]] != "-":
                    print(f"Место '{coordinate[0]} {coordinate[1]}' уже занято. Повторите ввод.")
                    return False
        else:
            print('Вы ввели неправельные координаты. Повторите ввод.')
            return False
    
    return coordinate


def insert_coordinates(coordinate, ch):
    """Данная функция заносит фигуру игрока на указанную позицию
    """
    
    if coordinate[0] == 1:
        L1[coordinate[1]] = ch
    elif coordinate[0] == 2:
        L2[coordinate[1]] = ch
    elif coordinate[0] == 3:
        L3[coordinate[1]] = ch


def check_victory(ch):
    """Данная функция определяет победителя
    """
    
    if L1[1] == L1[2] == L1[3] != "-" or L2[1] == L2[2] == L2[3] != "-"\
            or L3[1] == L3[2] == L3[3] != "-" or L1[1] == L2[2] == L3[3] != "-"\
            or L1[3] == L2[2] == L3[1] != "-" or L1[1] == L2[1] == L3[1] != "-"\
            or L1[2] == L2[2] == L3[2] != "-" or L1[3] == L2[3] == L3[3] != "-":
        print(f"Игрок '{ch}', Вы победили!!!")
        return True


def print_field():
    """Вывод поля на печать"""
    
    print('     ',' '.join(L0))
    print('     ',' '.join(L1))
    print('     ',' '.join(L2))
    print('     ',' '.join(L3))

    
# Определяем игроков    
player = 'X'
player_to_wait = 'O'


# Начало игры
print('   Приветсвуем вас в игре')
print('     "крестики и нолики" ')
print('----------------------------')
print('Диапозон координат от 1 до 3')
print()
print_field()


#Игровой процесс
while True:
    
    #ввод координат
    step = input(f"Введите координаты '{player}' строка - столбец через пробел: ")
    
    #проверка координат
    if error_coordinates(step):
        step = error_coordinates(step)
    else:
        continue
    
    #Занесение в игровое поле   
    insert_coordinates(step, player)
    print_field()
    
    #Проверка на победу и на ничью
    if check_victory(player):
        break
    if L1.count("-") == 0 and L2.count("-") == 0 and L3.count("-") == 0:
        print("Игра окончена. Ничья!!!")
        break
    
    #Смена хода
    player, player_to_wait = player_to_wait, player

```
