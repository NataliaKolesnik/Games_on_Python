```python
import random
import math

# Разовая игра - Угадай число


number = random.randint(1, 101)  # Типа загадали число

print("Загадано число от 1 до 100. Попробуй угадай :)")

count = 0  # счетчик попыток
i, j = 50, 25  # i - проверочное число, j - корректировка i


while True:
    count += 1

    if number == i:
        break
    elif number < i:
        i -= j
    else:
        i += j
    
    # округляем в большую сторону, чтоб не получилось значение j = 0 (при делении 1/2)
    j = math.ceil(j/2)

    
print(f"Вы угадали!!! за {count} попыток. Было загаданно число {number}")

```
